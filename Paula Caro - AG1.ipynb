## 1. Torres de Hanoi (Divide y Vencerás)

**Problema**

Es un juego matemático que consiste en tres postes verticales. En uno de los postes se apila un número de discos perforados por su centro, que determinará la complejidad de la solución. Los discos se apilan sobre uno de los postes en tamaño decreciente de abajo arriba, y el juego consiste en pasar todos los discos desde el poste ocupado a uno de los otros postes vaci¡íos.

**Reglas**
Para realizar este objetivo es necasario seguir tres simples reglas:
1. Solo se puede mover un disco cada vez y para mover otro los demás tienen que estar en postes.
2. Un disco de mayor tamaño no puede estar sobre uno más pequeño que él mismo.
3. Solo se puede desplazar el disco que se encuentre arriba en cada poste.

```python
def mover_ficha(torres, desde, hasta):
    ficha = torres[desde].pop()
    torres[hasta].append(ficha)
    print(f"Lleva la ficha {ficha} desde la Torre {desde} hasta la Torre {hasta}.")

def posiciones_actuales(torres):
    print(torres)

def Torres_Hanoi(N, desde, hasta, torres):  # Mejora: más dimensiones?
    if N ==1 :
        mover_ficha(torres, desde, hasta)

    else:
       # Torres_Hanoi(N-1, desde, 6 - desde - hasta )
       # El objetivo del primer "bucle" es mover las fichas hasta la columna pivote ( 6 -1 - 3 = 2)
       Torres_Hanoi(N - 1, desde, 6 - desde - hasta, torres)
       mover_ficha(torres, desde, hasta)

       #Torres_Hanoi(N-1,6-desde-hasta, hasta )
       # El objetivo del segundo "bucle" es mover las fichas desde la pivote hasta la destino
       Torres_Hanoi(N-1, 6-desde-hasta  , hasta, torres)

torres = {
    1: ["Verde", "Amarilla", "Roja"],
    2: [],
    3: []
    }
posiciones_actuales(torres)
Torres_Hanoi(3, 1 , 3, torres)
posiciones_actuales(torres)
```

### **Mejora Propuesta: *Torres de Hanoi Multidimensional***

```python
import math

# --- 1. Diccionario ---
torres = {}

def imprimir_estado():
    # Función auxiliar para ver las torres
    lista_torres = sorted(torres.keys())
    print("\nEstado actual:")
    for t in lista_torres:
        print(f" {t}: {torres[t]}")
    print("-" * 30)

def mover_ficha(desde, hasta):
    ficha = torres[desde].pop()
    torres[hasta].append(ficha)
    print(f" >> Mover disco {ficha} | {desde} -> {hasta}")

def hanoi_multidimensional(n, origen, destino, auxiliares):
    """
    Algoritmo Frame-Stewart para K torres.
        n:          número de discos a mover
        origen:     clave de la torre origen (ej: 'A')
        destino:    clave de la torre destino (ej: 'C')
        auxiliares: LISTA de claves de torres disponibles como auxiliares
    """
    
    # --- Caso Base: 0 discos ---
    if n == 0:
        return
    
    # --- Caso Base: 1 disco ---
    if n == 1:
        mover_ficha(origen, destino)
        return

    # --- Caso Estándar: 3 Torres (Solo 1 auxiliar disponible) ---
    # Si solo queda 1 auxiliar, usamos el algoritmo clásico O(2^n)
    if len(auxiliares) == 1:
        aux = auxiliares[0]
        hanoi_multidimensional(n - 1, origen, aux, [destino])
        mover_ficha(origen, destino)
        hanoi_multidimensional(n - 1, aux, destino, [origen])
        return

    # --- Caso Multidimensional: K > 3 ---
    # Algoritmo Frame-Stewart: Dividimos la torre en dos montones.
    # Montón superior (k discos) y Montón inferior (n-k discos).
    
    # Cálculo heurístico de k (tamaño del montón superior)
    # k ≈ n - sqrt(2n) es una aproximación eficiente.
    k = n - round(math.sqrt(2 * n))
    if k < 1: k = 1  # Seguridad para n pequeños
    
    # Elegimos una torre de las auxiliares para dejar el montón 'k' temporalmente
    torre_pivote = auxiliares[0]
    resto_auxiliares = auxiliares[1:] # Las demás torres libres

    # PASO 1: Mover los k discos superiores a la torre pivote
    # Usamos todas las torres disponibles (incluido el destino final) como ayuda
    hanoi_multidimensional(k, origen, torre_pivote, resto_auxiliares + [destino])

    # PASO 2: Mover los (n-k) discos inferiores al destino
    # Aquí la torre_pivote no se puede usar (está ocupada con los k discos)
    hanoi_multidimensional(n - k, origen, destino, resto_auxiliares)

    # PASO 3: Mover los k discos desde la pivote al destino final
    # Usamos todas las torres disponibles (incluido el origen original) como ayuda
    hanoi_multidimensional(k, torre_pivote, destino, resto_auxiliares + [origen])


#--------------- FUNCIÓN PRINCIPAL ---------------#

def iniciar_juego(num_discos, num_torres):
    global torres
    torres = {}
    
    # Generamos las claves de las torres(A, B, C,...)
    claves_torres = [chr(65 + i) for i in range(num_torres)]
    
    # Llenamos la primera torre (A) y dejamos vacías las demás
    torres[claves_torres[0]] = list(range(num_discos, 0, -1))
    for k in claves_torres[1:]:
        torres[k] = []
        
    origen = claves_torres[0]      
    destino = claves_torres[-1]    

    # Las auxiliares son todas las que no son ni origen ni destino
    auxiliares = claves_torres[1:-1] 

    print(f"--- Iniciando Hanoi con {num_discos} discos y {num_torres} torres ---")
    print(f"Origen: {origen}, Destino: {destino}, Auxiliares: {auxiliares}")
    imprimir_estado()
    
    # Llamada inicial
    hanoi_multidimensional(num_discos, origen, destino, auxiliares)
    
    print("\n--- ¡Juego Terminado! ---")
    imprimir_estado()

# --- PRUEBA ---
# Probamos con 4 torres y 5 discos para ver la diferencia
iniciar_juego(num_discos=6, num_torres=4)
```

## 3. Devolución de cambio por técnica voraz

**Problema: Devolver cambio de monedas.**
- Dado un sistema monetario $[1, v_1, v_2, \dots, v_n]$ descomponer cualquier cantidad $C$ de manera que usemos el menor número de monedas.

```python
def cambio_monedas(N, SM):
  """
  El sistema monetario debe estar ordenado de mayor a menor.
  """
  SOLUCION = [0]*len(SM)   # SOLUCION = [0,0,0,0,..]
  ValorAcumulado = 0

  for i, valor in enumerate(SM):
    monedas =  (N - ValorAcumulado) // valor
    SOLUCION[i] = monedas
    ValorAcumulado = ValorAcumulado + monedas * valor

    if ValorAcumulado == N:
      return SOLUCION


print(cambio_monedas(15,[25,10,5,1]))
print(cambio_monedas(15, [11, 5, 1]))
```

## 4. N-Reinas por técnica de vuelta atrás

- **Planteamiento del problema:** Colocar en un tablero de tamaño $n\times n$, $n$ reinas sin que se amenacen.
- **¿Cómo lo modelamos?**
    - Tendremos $n$-tuplas $(x_1, x_2, \dots, x_n)$ donde el valor de cada elemento es la fila donde está posicionada la reina en la columna $i$-ésima.
    - El **árbol de expansión** recorrerá todas las posibilidades. Así, es posible determinar si una solución parcial (rama del árbol) es "prometedora" / factible.
    - Las restricciones a tener en cuenta son:
        - No puede haber dos reinas en la misma **columna** $\to$ esta restricción se verifica por el modelo que hemos adoptado.
        - No puede haber dos reinas en la misma **fila** $\to$ no puede haber dos valores iguales para una solución parcial.
        - No puede haber dos reinas en la misma **diagonal** $\to$ $| x_i - x_j | \neq | i - j|$

```python
def escribe(S): # Pinta el tablero de ajedrez
  n = len(S)
  for x in range(n):
    print("")
    for i in range(n):
      if S[i] == x+1:
        print(" X " , end="")
      else:
        print(" - ", end="")


def es_prometedora(SOLUCION,etapa):
  # print("********************")
  # print(SOLUCION)
  # print(escribe(SOLUCION))
  # print("********************\n")

  #Si la solución tiene dos valores iguales no es valida => Dos reinas en la misma fila
  for i in range(etapa+1):
    #print("El valor " + str(SOLUCION[i]) + " está " +  str(SOLUCION.count(SOLUCION[i])) + " veces")
    if SOLUCION.count(SOLUCION[i]) > 1: # Algo de la forma (0, 2, 2, 2)
      return False

    #Verifica las diagonales
    for j in range(i+1, etapa +1 ):
      #print("Comprobando diagonal de " + str(i) + " y " + str(j))
      if abs(i-j) == abs(SOLUCION[i]-SOLUCION[j]) : return False # Ej: (0, 2, 1, 3) (True) // (0, 1, 2, 3) (False)
  return True



def reinas(N, solucion=[], etapa=0):
  if len(solucion) == 0:
      solucion=[0 for i in range(N)]

  for i in range(1, N+1):
    solucion[etapa] = i

    if es_prometedora(solucion, etapa):
      if etapa == N-1:
        print(solucion)
        escribe(solucion)
        print()
      else:
        reinas(N, solucion, etapa+1)
    else:
      None

    solucion[etapa] = 0

reinas(4)
```

## 5. Viaje por el rio. Programación dinámica

*Programación Dinámica*: similar al divide y vencerás, pero la diferencia es que aprovechamos soluciones intermedias para llegar a la solución final.

**Problema: Viaje por el rio**
- En un río hay $n$ embarcaderos y debemos desplazarnos río abajo desde un embarcadero a otro. Cada embarcadero tiene precios diferntes para ir de un embarcadero a otro situado más abajo. Para ir del embarcadero $i$ al $j$, puede ocurrir que sea más barato hacer un trasbordo por un embarcadero intermedio $k$. El problema consiste en determinar la combinación más barata.
- ***Restricciones:*** No hay posibilidad de remontar el río. Siempre debemos ir río abajo.
- ***Modelado:*** 
    - Consideramos una tabla $T(i, j)$ para almacenar todos los precios que nos ofrecen los embarcaderos.
    - Si no es posible ir desde $i$ a $j$ daremos un valor alto para garantizar que ese trayecto no se va a elegir en la ruta óptima.
    - Establecer una tabla intermedia para...
    $$C_{i, j} = \min\{T_{i, j}, T_{i, k} + C_{k, j} \quad | \quad  \forall i < k \leq j \}$$

```python
TARIFAS = [
  [0  ,5  ,4  ,3  ,999,999,999],
  [999,0  ,999,2  ,3  ,999,11 ],
  [999,999,0  ,1  ,999,4  ,10 ],
  [999,999,999,0  ,5  ,6  ,9  ],
  [999,999,999,999,0  ,999,4  ],
  [999,999,999,999,999,0  ,3  ],
  [999,999,999,999,999,999,0  ]
]



################################################################
def Precios(TARIFAS):
  # Para cada una de las rutas, calcula cuanto acabaríamos pagando.
################################################################
  #Total de Nodos
  N = len(TARIFAS[0])

  #Inicialización de la tabla de precios
  PRECIOS = [ [9999]*N for i in [9999]*N]
  RUTA = [ [""]*N for i in [""]*N]

  for i in range(0,N-1):
    RUTA[i][i] = i             #Para ir de i a i se "pasa por i"
    PRECIOS[i][i] = 0          #Para ir de i a i se se paga 0
    for j in range(i+1, N):
      MIN = TARIFAS[i][j]
      RUTA[i][j] = i

      for k in range(i, j):
        if PRECIOS[i][k] + TARIFAS[k][j] < MIN:
            MIN = min(MIN, PRECIOS[i][k] + TARIFAS[k][j] )
            RUTA[i][j] = k          #Anota que para ir de i a j hay que pasar por k
        PRECIOS[i][j] = MIN

  return PRECIOS,RUTA
################################################################


PRECIOS,RUTA = Precios(TARIFAS)
#print(PRECIOS[0][6])

print("PRECIOS")
for i in range(len(TARIFAS)):
  print(PRECIOS[i])

print("\nRUTA")
for i in range(len(TARIFAS)):
  print(RUTA[i])

#Determinar la ruta con Recursividad
def calcular_ruta(RUTA, desde, hasta):
  if desde == hasta:
    #print("Ir a :" + str(desde))
    return ""
  else:
    return str(calcular_ruta( RUTA, desde, RUTA[desde][hasta])) +  \
                ',' + \
                str(RUTA[desde][hasta] \
              )

print("\nLa ruta es:")
calcular_ruta(RUTA, 0,6)
```
